##### 1) Instruction ( For Data Cleaning )
##### Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.


```python
cardata.isnull().sum()
```

NOTE: Here all columns have null value to replace null value in all numerical columns use cardata.mean() and
to replace null value in all non-numerical columns use cardata.mode()

```python
# lis=list(cardata.columns)
# lis
meanValue = cardata.mean(axis=0,numeric_only=True,skipna=True)
meanValue
```

```python
modeValue= cardata.mode().iloc[0]
modeValue
```

```python
# meanValue=cardata.mean() # ---> FutureWarning
# meanValue            
```

```python
# see all null value replaced
cardata.fillna(value = meanValue, inplace=True)
cardata.fillna(value= modeValue, inplace=True)
```

```python
cardata.isnull().sum()
```

##### 2) Problem ( Based on Value Counts )- Check what are the different types of Make are there in our dataset. And, what is the count (occurrence) of each Make in the data ?


```python
# different types of Make
cardata['Make'].unique()
```

```python
# Total number of distinct types of Make
cardata['Make'].value_counts().count()
```

```python
# count of occurrence of each Make in the dataset
cardata['Make'].value_counts()
```

##### 3) Instruction ( Filtering ) 
##### Show all the records where Origin is Asia or Europe.


 * isin() method helps in selecting rows with having a particular(or Multiple) value in a particular column.

```python
cardata[cardata['Origin'].isin(['Asia', 'Europe'])]
```

##### 4) Instruction ( Removing unwanted records )
##### Remove all the records (rows) where Weight is above 4000.


* To remove rows you can use these three Ways.

```python
#1
cardata[~(cardata['Weight']>4000)]
```

```python
#2
indexValue=cardata[cardata['Weight']>4000].index
cardata.drop(indexValue, inplace=True)
```

```python
#3
cardata.drop(cardata.index[cardata['Weight']>4000], inplace= True)
```

```python
# Now you can check Greater then 4000 all value are removed
# before it was (432,15) and now
cardata.shape
```

##### 5) Instruction ( Applying function on a column )
##### Increase all the values of 'MPG_City' column by 3.


```python
cardata['MPG_City'] = cardata['MPG_City'].apply(lambda x:x+3)
```
